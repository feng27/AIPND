```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Read csv
pokemon = pd.read_csv('./data/pokemon.csv')
print(pokemon.shape)
pokemon.head()
```

```python
sorted_counts = pokemon['generation_id'].value_counts()
#sorted_counts
plt.pie(sorted_counts, labels= sorted_counts.index, startangle = 180, counterclock = False);

# We have the used option `Square`. 
# Though, you can use either one specified here - https://matplotlib.org/api/_as_gen/matplotlib.pyplot.axis.html?highlight=pyplot%20axis#matplotlib-pyplot-axis
plt.axis('square');
```

### add label, percentage, shift

```python
pie_labels = ['gen{}'.format(v) for v in sorted_counts.index]
pie_labels
pie_explode = np.zeros(sorted_counts.shape[0])
pie_explode[0]=0.1
plt.pie(sorted_counts, labels=pie_labels, startangle=90, counterclock=False, autopct='%1.1f%%', explode=pie_explode, shadow=True)
plt.axis('square');
```

### donut plot

```python
plt.pie(sorted_counts, labels=pie_labels, startangle=90, counterclock=False, wedgeprops={'width': 0.4})
plt.axis('square');
```
